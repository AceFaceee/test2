```python
#筛选B列为14的数据集
filtered = df.loc[df['_zJpxzcA_zIcx']==14]
filtered
```

```python
from sklearn.preprocessing import PolynomialFeatures

x = filtered_qua[['_zgJcx','xIIzYAcC7p_HW','_zgJcxp','_zgJcIp']]
y = filtered_qua['gCHciHxxcxGYcNDEcK7D']

#提取二阶特征值 ScalerFeatures
po = PolynomialFeatures(degree=2, interaction_only=False, include_bias=False)
x_poly = po.fit_transform(x)
pd.DataFrame(x_poly).head()
```

```python
#贴标签
#x1 = '_zgJcx'
#x2 = 'xIIzYAcC7p_HW'
#x3 = '_zgJcxp'
#x4 = '_zgJcIp'
x_poly = pd.DataFrame(x_poly,columns=['x1','x2','x3','x4','x1^2',\
                                      'x1_x2','x1_x3',\
                                     'x1_x4','x2^2','x2_x3','x2_x4',\
                                     'x3^2','x3_x4','x4^2'])
x_poly.head()
```

```python
#拆分训练集和测试集
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import Lasso,Ridge,LinearRegression as LR
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import r2_score, explained_variance_score as \
EVS,mean_squared_error as MSE

x_train,x_test,y_train,y_test = train_test_split(x_poly,y,test_size=0.2\
                                                ,random_state=1)
```

```python
polyReg = LR().fit(x_train,y_train)
yHat = polyReg.predict(x_test)
yHat
```

```python
polyReg.coef_
```

```python
polyReg.intercept_
```

```python
#计算RMSE
np.sqrt(MSE(y_test,yHat))/y_test.mean()
```

```python
#计算R^2
r2_score(y_pred=yHat, y_true=y_test)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import lstsq

# 2. 尝试求解 以及 可视化
fig = plt.figure(figsize=(10, 5))
m = np.ones(x_1.shape)
m = np.hstack((m, x_1, x_1 ** 2, x_2, x_2 ** 2, x_1*x_2))
k = lstsq(m, y_t, rcond=None)[0].reshape(-1)
print(k)
# 可视化
ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.scatter(x_1.reshape(-1), x_2.reshape(-1), y_t.reshape(-1), c='red', s=20, label='标签')
x1 = np.linspace(0, 1, 100)
x2 = np.linspace(0, 1, 100)
x, y =np.meshgrid(x1, x2)
z = k[0] + k[1] * x + k[2] * x ** 2 + k[3] * y + k[4] * y**2 + k[5] * x * y
ax.plot_surface(x, y, z,rstride=4,cstride=4,alpha=0.6)
ax.legend()
ax.set_title('二元二次多项式归回')
ax.view_init(12,-78)
```
